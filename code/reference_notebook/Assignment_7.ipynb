```python
import numpy as np
import matplotlib.pyplot as plt
```

**External-Influence Model**

```python
t = np.arange(5000)
alpha = [0.01, 0.001, 0.0001]

P = 1000
A = []

for i in range(3):
    res = P *(1-np.exp(-alpha[i]*t))
    A.append(res)

ax = plt.axes()
ax.plot(t, A[0], 'r', label='alpha=0.01')
ax.plot(t, A[1], 'g', label='alpha=0.001')
ax.plot(t, A[2], 'b', label='alpha=0.0001')
ax.set_title('External-Influence Model')
ax.set_xlabel('t')
ax.set_ylabel('A(t)')
ax.set_ylim(0, 1200)
ax.legend()
plt.show()
```

**Internal-Influence Model**

```python
t = np.arange(2500)
beta = [0.001, 0.0001, 0.00001]
A0=50

P = 1000

A = []
for i in range(3):
    res = P/(1+(P-A0)/A0*np.exp(-beta[i]*P*t))
    A.append(res)

ax = plt.axes()
ax.plot(t, A[0], 'r', label='beta1=0.001')
ax.plot(t, A[1], 'g', label='beta2=0.0001')
ax.plot(t, A[2], 'b', label='beta3=0.00001')
ax.set_title('Internal-Influence Model')
ax.set_xlabel('t')
ax.set_ylabel('A(t)')
ax.set_ylim(0, 1200)
ax.legend()
plt.show()
```

**Mixed-Influence Model**

```python
t = np.arange(3000)
alpha = [0.01, 0.001, 0.0001]
beta = [0.001, 0.0001, 0.00001]
A0=50
P = 1000
A = []
for i in range(3):
    up = (P-alpha[i]*(P-A0)/(alpha[i]+beta[i]*A0)*np.exp(-(alpha[i]+beta[i]*P)*t))
    down = (1+beta[i]*(P-A0)/(alpha[i]+beta[i]*A0)*np.exp(-(alpha[i]+beta[i]*P)*t))
    res = up / down
    A.append(res)

ax = plt.axes()
ax.plot(t, A[0], 'r', label='alpha1= 0.01, beta1=0.001')
ax.plot(t, A[1], 'g', label='alpha2= 0.001, beta1=0.0001')
ax.plot(t, A[2], 'b', label='alpha3= 0.0001, beta1=0.00001')
ax.set_title('mixed-Influence Model')
ax.set_xlabel('t')
ax.set_ylabel('A(t)')
ax.set_ylim(0, 1200)
ax.legend()
plt.show()
```

**SIRS Model**

$\frac{dS}{dt} = -\beta  S I + \xi R $<br>
$\frac{dI}{dt} = \beta  S I - (\gamma )I $<br>
$\frac{dR}{dt} = \gamma I - \xi R $<br>

$S + I + R = 1$

**SEIRS Model**

$\frac{dS}{dt} = -\beta  S E + \xi R $<br>
$\frac{dE}{dt} = \beta  S E - \sigma E $<br>
$\frac{dI}{dt} = \sigma E - \gamma I $<br>
$\frac{dR}{dt} = \gamma I - \xi R $<br>

$S + E + I + R = 1$
